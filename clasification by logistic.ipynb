```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
sns.set()
```

```python
titanic=pd.read_csv('titanic_data.csv')
```

```python
titanic.head()
```

```python
titanic.describe(include='all')
```

```python
newtitanic = titanic.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1)
```

```python
newtitanic
# i have removed all unwanted features from data as they are not a factor in survival
```

```python
newtitanic.describe(include='all')
#here i have found some missing value in Age columns show i will filled it my average age
```

```python
newtitanic.isnull().sum()
```

```python
clean_titanic = newtitanic.fillna(29)
```

```python
clean_titanic
# i will now convert catogorical data which is sex in numerical data
```

```python
clean_data=clean_titanic.copy()
clean_data['Sex']=clean_data['Sex'].map({'female':0 ,'male':1})
clean_data.head(10)
```

```python
x=clean_data.iloc[:,[1,2,3]]
y=clean_data.iloc[:,[0]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_Train,x_Test,y_Train,y_Test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clsf= LogisticRegression()
```

```python
clsf.fit(x_Train,y_Train)
```

```python
survival_predict=clsf.predict(x_Test)
survival_predict
```

```python
y_Test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_Test,survival_predict)
cm
# here is the confusion matrix that show how my model is good in prediction
# here i am clearly seeing that 72% chance of surviaval is explained by model. 
# survival of person also depend on many things so my model doent have those features persons physical condition and etc
```
